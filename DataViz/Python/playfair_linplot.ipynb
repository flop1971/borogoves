```python
# Declare Libaries
import pandas as pd
import re
import sys
import datetime
import numpy as np
import bokeh
from bokeh.layouts import gridplot
from bokeh.models import Label
from bokeh.plotting import figure, show, output_file, output_notebook
from bokeh.palettes import Spectral11, colorblind, Inferno, BuGn, brewer
from bokeh.models import HoverTool, value, LabelSet, Legend, ColumnDataSource,LinearColorMapper,BasicTicker, PrintfTickFormatter, ColorBar
```

```python
# Import dataframe to a csv
df_playfair = pd.read_csv ('playfair-estimated.csv')



df_playfair

```

```python
col_names = list(df_playfair.columns)
col_1 = (col_names[1:2]) 
col_2 = (col_names[2:3]) 

def converttostr(input_seq, seperator):
   # Join all the strings in list
   final_str = seperator.join(input_seq)
   return final_str

seperator = ' '
col1_string =  converttostr(col_1, seperator)
col2_string =  converttostr(col_2, seperator)

print(col1_string)
print(col2_string)
```

```python
# tools preamble
TOOLS = 'save,pan,box_zoom,reset,wheel_zoom,hover'

df_playfair['Year'] = pd.to_datetime(df_playfair['Year'])

# chart Aspect Ratio 
ar_line = 1.75
plot_high = 700
plot_wd = plot_high*ar_line

int_plot_wd = int(plot_high*ar_line)



# Colour list = Dark Blue, Orange, Brown, Light Grey, Dark Grey
protan_5= ["#3a6587", "#e59216", "#964B00", "#aeb3b7", "7F7F7F"]

#source = ColumnDataSource(data=dict(height=[192,87],
#                                    weight=[500,500],
#                                    names=['Exports', 'Imports']))




# Plot the values as a line chart
p1 = figure(title="Imports to and from Denmark & Norway from 1700 to 1780", 
            x_axis_type="datetime", 
            plot_height = plot_high,
            tools = TOOLS, 
            plot_width = int_plot_wd)

p1.line(x=df_playfair['Year'].tolist(), 
        y=df_playfair.Exports, 
        line_color="#3a6587", 
        line_width = 3)

p1.line(x=df_playfair['Year'].tolist(), 
        y=df_playfair.Imports, 
        line_color="#aeb3b7", 
        line_width = 3)



# Add legend items with Gestalt Similarity

p1.add_layout(Label(x=-6005814400000, y=188, text='Exports',text_color=protan_5[0]))
p1.add_layout(Label(x=-6005814400000, y=93, text='Imports',text_color=protan_5[3]))



p1.select_one(HoverTool).tooltips = [
    ('Pounds Sterling', '@x')
]

# Chart Aesthetics

# Mute the chart gridlines (i.e.. removes the chart clutter)
p1.xgrid.grid_line_color = "#aeb3b7"
p1.xgrid.grid_line_dash = "dotted"
p1.xgrid.grid_line_alpha = 0.5

p1.ygrid.grid_line_color = "#aeb3b7"
p1.ygrid.grid_line_dash = "dotted"
p1.xgrid.grid_line_alpha = 0.5

# change just some things about the x-axes
p1.xaxis.axis_label = "Date / Time"
p1.xaxis.axis_line_width = 2
p1.xaxis.major_label_text_color = "black"
p1.xaxis.axis_line_color = "#aeb3b7"

# change just some things about the y-axes
p1.yaxis.axis_label = "Pounds (£)"
p1.yaxis.axis_line_width = 2
p1.yaxis.major_label_text_color = "black"
p1.yaxis.axis_line_color = "#aeb3b7"
p1.yaxis.major_label_orientation = "vertical"

# Set the range of the chart (can overide figure settings)
p1.x_range.start = -8520336000000
p1.y_range.start = 0
p1.y_range.end = df_playfair.Exports.max() + (df_playfair.Exports.max()*0.1)

# Remove the border. Set the width to 0 does not work so we need
# to set to 0.1 to make it less visible.
p1.outline_line_width = 0.1

# Set attributes for the chart title
p1.title.text_color = "black"
p1.title.text_font_size = "20pt"
#p.title.text_font = "times"
#p.title.text_font_style = "italic"
p1.title.align = "center"

# Mute the Gridlines



p1.legend.location = "top_right"
# Rotate the orientation of the y label data
# Don't set to 0 as it will encroach on the chart columns
# Setting to a value close to 0 would show an almost horitzonal
# Category
p1.yaxis.major_label_orientation = 0.0001


show(p1)
```
